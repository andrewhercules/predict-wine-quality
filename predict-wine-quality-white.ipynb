## Clean data

```python
# split quality ratings into categorical bins - label in parenthesis:
# 1 - 4 = poor (0)
# 5 - 6 = average (1)
# 7 - 10 = good (2)

bins = [1,4,6,10]

quality_labels=[0,1,2]

white_wine_data['quality_category'] = pd.cut(white_wine_data['quality'], bins=bins, labels=quality_labels, include_lowest=True)

white_wine_data['quality_category'].value_counts()
```

## Build prediction models

```python
# import relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import scale
```

```python
# set features and labels (wine_quality)
features = white_wine_data.drop(['quality', 'quality_category'], axis=1)
scaled_features = scale(features, axis=0)
wine_quality = white_wine_data['quality_category']

# split data into training and test sets
train_data, test_data, train_labels, test_labels = train_test_split(scaled_features, wine_quality, test_size=0.2, random_state=25)
```

```python
# create list to store data about the performance of each model
results = []

# create list of models
models = [
    SVC(),
    DecisionTreeClassifier(),
    KNeighborsClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    GaussianNB()
]

# iterate through list of models and run predictions and store accuracy values
for model in models:
    result = {}
    classifier_name = model.__class__.__name__
    kfold = KFold(n_splits=10, random_state=25)
    cross_val_results = cross_val_score(model, train_data, train_labels, cv=kfold, scoring='accuracy')
    result['classifier_name'] = classifier_name
    result['classifier_results'] = cross_val_results
    results_message = "Accuracy of %s: %0.2f (+/- %0.2f)" % (classifier_name, cross_val_results.mean(), cross_val_results.std())
    print(results_message)
```

```python
# create RandomForest Classifier and fit model with training data and labels
rf_classifier = RandomForestClassifier()
rf_classifier.fit(train_data, train_labels)

# create list of feature labels
feature_labels = list(white_wine_data.columns.values)

# generate list of feature importance
print('*** Feature importance ***')
for feature in zip(feature_labels, rf_classifier.feature_importances_):
    print(feature)

# import feature selection library
from sklearn.feature_selection import SelectFromModel

# create selector that will use the Random Forest classifer to identify features with an importance greater than the mean of all importances

sfm = SelectFromModel(rf_classifier, threshold='mean')
sfm.fit(train_data, train_labels)

print('*** Features with importance of more than the mean of all features ***')    
for feature_list_index in sfm.get_support(indices=True):
    print(feature_labels[feature_list_index])
```

```python
train_data_important_features = sfm.transform(train_data)
test_data_important_features = sfm.transform(test_data)

# Create a new random forest classifier for the most important features
rf_classifier_important_features = RandomForestClassifier()

rf_classifier_important_features.fit(train_data_important_features, train_labels)

from sklearn.metrics import accuracy_score

predictions_all_features = rf_classifier.predict(test_data)

accuracy_score(test_labels, predictions_all_features)
```

```python
predictions_important_features = rf_classifier_important_features.predict(test_data_important_features)

accuracy_score(test_labels, predictions_important_features)
```
